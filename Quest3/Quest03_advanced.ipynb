````python
from google.colab import drive
drive.mount('/content/drive')

# 기존 방식은 런타임 재실행 마다 파일을 새로 업로드해주어야 해서 드라이브 마운트 방식으로 변경
````

````python
import string

file_line = []
with open('/content/drive/MyDrive/aiffel_python_quest/06TheAvengers.txt', 'r') as file:
  for line in file:
    temp = line.strip('\n').casefold()
    for i in string.punctuation:
      temp = temp.replace(i, ' ')
    # temp = temp.translate(str.maketrans('', '', string.punctuation))
    # temp = ''.join(filter(lambda x: x not in string.punctuation, temp))
    
    # 기존 방식으로 기호를 제거했더니 기호 앞뒤 문자가 붙어버려 아쉬움이 있었음
    # 기호 대신 공백을 replace 하는 방법으로 전처리
    
    file_line.append(temp)
file_line 
````

````python
string.punctuation
````

````python
baglist = []
# 줄 단위로 불러와서 공백을 기준으로 단어를 나눈 words 리스트를 생성하고
# 두 단어씩 묶은 2-gram 튜플을 gram_temp에 저장해서 baglist에 더함
for line in file_line:
  words = line.split()
  for i in range(len(words) - 1):
    gram_temp = (words[i], words[i + 1])
    baglist.append(gram_temp)
baglist
````

````python
from collections import Counter
#baglist에 있는 요소들의 갯수를 세주는 클래스 Counter 사용
countdict = Counter(baglist)

# 딕셔너리의 value 값을 기준으로 내림차순으로 정렬
sorted_countdict = sorted(countdict.items(), key=lambda x: x[1], reverse=True)
sorted_countdict 
````

````python
# 내림차순으로 정렬된 리스트에서 가장 첫 번째 값 = 가장 큰 값
max2gram = sorted_countdict[0][0]
print(max2gram, countdict[max2gram])
print(Counter(baglist))
````
